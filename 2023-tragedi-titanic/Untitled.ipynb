```python
import pandas as pd

def summary(df):
    data = pd.DataFrame(index=df.columns)
    data['dtypes'] = df.dtypes
    data['missing'] = df.isna().sum()
    data['%missing'] = df.isna().sum() / len(df) * 100
    data['value_counts'] = df.count()

    # Create a summary DataFrame with statistics using describe()
    desc_stats = df.describe().transpose()
    data = pd.concat([data, desc_stats], axis=1)

    return data

# Example usage:

```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score, precision_score

# Assuming xtrain, ytrain, xtest, ytest are your training and testing datasets

# Define the pipeline
step1 = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(sparse=False, drop='first', handle_unknown='ignore'), [1, 6, 7])
], remainder='passthrough')

step3 = RandomForestClassifier()

pipe10 = Pipeline([
    ('step1', step1),
    # ('step2', StandardScaler()),  # You can uncomment and include StandardScaler if needed
    ('step3', step3)
])

# Define the parameter grid for RandomForestClassifier
param_grid = {
    'step3__n_estimators': [50, 100, 200],
    'step3__max_depth': [None, 10, 20],
    'step3__min_samples_split': [2, 5, 10],
    'step3__min_samples_leaf': [1, 2, 4]
}

# Use GridSearchCV for parameter tuning
grid_search = GridSearchCV(pipe10, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(xtrian, ytrian)

# Get the best parameters and the best model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Predict using the best model
y_pred = best_model.predict(xtest)

# Evaluate the performance
print('Best Parameters:', best_params)
print('Accuracy:', grid_search.best_score_)
print('R2 Score:', r2_score(ytest, y_pred))
print('Precision Score:', precision_score(ytest, y_pred))
```
